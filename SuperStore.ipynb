## Data Understanding

* Data of Retail Transaction from 01 Januari 2017 to 31 December 2020

* Source Data: SuperStore.
* The dataset has 20 columns and 9994 rows.
* Data Dictionary:
* Order_ID: Unique identification number for each order placed by a customer.
* Customer_ID: Unique identification number for each customer.
* Postal_Code: The postal code where the customer is located.
* Product_ID: Unique identification number for each product.
* Sales: The total number of sales for a product in an order.
* Quantity: The number of product units purchased in an order.
* Discount: Discount given on a product in percentage.
* Profit: The profit generated from the sale of products in an order.
* Category: Product category, for example, "Technology," "Clothing," etc.
* Sub-Category: A product sub-category, provides more details about the types of products in a particular category.
* Product_Name: Name of the product being sold.
* Order_Date: The date when the order was created.
* Ship_Date: Order delivery date.
* Ship_Mode: The shipping method used to send orders to customers.
* Customer_Name: Name of the customer who placed the order.
* Segment: Customer segment, for example, "Consumer," "Corporate," "Home Office."
* Country/Region: The country or region where the customer is located.
* City: The name of the city where the customer is located.
* State: The name of the state or province where the customer is located.
* Region: The geographic region where the customer is located.

## Exploratory Data Analysis

### Sales Analysis

#### How do sales trend every year?


```python
#grouping sales by year
year_sales = df.groupby(df['Order_Date'].dt.year)['Sales','Profit'].sum()

#indexing
year_sales = year_sales.reset_index()
year_sales = year_sales.rename(columns={'Order_Date':'Year','Sales':'Total_Sales'})

year_sales
```

```python
#bar plot
plt.figure(figsize=(6,3))

#bar plot total sales by year
sns.barplot(
  y="Total_Sales",
  x="Year",
  data=year_sales.sort_values(by="Total_Sales", ascending=False),
  color='#FFA559',
  label='Total Sales'
)
# Bar plot Profit
sns.barplot(
    x="Year",
    y="Profit",
    data=year_sales.sort_values(by="Total_Sales", ascending=False),
    color='#FF6000',  # Adjust color as needed
    label='Profit'
)

plt.title("Total Sales per Year", loc="center", fontsize=15)
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Add data labels for Total Sales
for i, value in enumerate(year_sales.sort_values(by="Year", ascending=True)['Total_Sales']):
    plt.text(i, value, f'{value/1e3:.2f}K', ha='center', fontsize=10, color='black')

# Add data labels for Profit
for i, value in enumerate(year_sales.sort_values(by="Year", ascending=True)['Profit']):
    plt.text(i, value, f'{value/1e3:.2f}K', ha='center', va='top', fontsize=10, color='black')


plt.ylabel("Total Sales")
plt.legend()
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

The highest sales occurred in 2020 with sales of 733.22K or 32% of total sales with a profit of 93.44k

```python
#Sales Performance
monthly_sales = df.groupby(df['Order_Date'].dt.strftime('%Y-%m'))['Sales','Profit','Quantity'].sum()

#indexing
monthly_sales = monthly_sales.reset_index()
monthly_sales = monthly_sales.rename(columns={'Order_Date':'Date','Sales':'Total_Sales'})
monthly_sales
```

```python
#bar plot
plt.figure(figsize=(12,3))

#bar plot total sales
ax1 = sns.lineplot(
  y="Quantity",
  x="Date",
  data=monthly_sales,
  color='#38419D',
  label='Order',
  marker='o'
)

plt.title("Order Trend 2017-2020", loc="center", fontsize=15)
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Add data labels for Profit
#for i, value in enumerate(monthly_sales.sort_values(by="Date", ascending=True)['Total_Sales']):
  #  plt.text(i, value, f'{value/1000:.2f}K', ha='center', fontsize=10, color='#200E3A')

ax1.set_ylabel("Order Counts")

plt.ylabel("Total Order")
plt.legend()
plt.tick_params(axis='x', labelsize=10, rotation=90)

plt.show()
```

```python
#bar plot
plt.figure(figsize=(12,3))

#bar plot total sales
ax1 = sns.lineplot(
  y="Total_Sales",
  x="Date",
  data=monthly_sales,
  color='#38419D',
  label='Profit',
  marker='o'
)

plt.title("Sales Trend 2017-2020", loc="center", fontsize=15)
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Add data labels for Profit
#for i, value in enumerate(monthly_sales.sort_values(by="Date", ascending=True)['Total_Sales']):
  #  plt.text(i, value, f'{value/1000:.2f}K', ha='center', fontsize=10, color='#200E3A')

ax1.set_ylabel("Sales")

plt.ylabel("Total Sales")
plt.legend()
plt.tick_params(axis='x', labelsize=10, rotation=90)

plt.show()
```

The sales trend looks quite fluctuating. The highest sales occurred in November 2020 with sales of 118.45k.Looking closer reveals a pattern where sales are typically quite high in September and November of each year.

#### How do profits perform every year?


```python
#bar plot
plt.figure(figsize=(12,3))

#bar plot total sales
ax1 = sns.lineplot(
  y="Profit",
  x="Date",
  data=monthly_sales,
  color='#38419D',
  label='Profit',
  marker='o'
)

plt.title("Profit Trend 2017-2020", loc="center", fontsize=15)
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

#Add data labels for Profit
#for i, value in enumerate(monthly_sales.sort_values(by="Date", ascending=True)['Profit']):
    #plt.text(i, value, f'{value/1000:.2f}K', ha='center', fontsize=10, color='#200E3A')

ax1.set_ylabel("Profit")

plt.ylabel("Total Profit")
plt.legend()
plt.tick_params(axis='x', labelsize=10, rotation=90)

plt.show()
```

Meanwhile, a profit of $17.89k in December 2019 was the highest profit.

#### How sales grow every year ?

```python
# Menghitung pertumbuhan penjualan dari satu bulan ke bulan lainnya
monthly_sales['Sales_Growth'] = monthly_sales['Total_Sales'].pct_change() * 100

# Membuat bar plot pertumbuhan penjualan
plt.figure(figsize=(10, 6))
bars = plt.bar(monthly_sales['Date'], monthly_sales['Sales_Growth'], color='#3887BE', label='Pertumbuhan Penjualan')

# Menambahkan line plot
line = plt.plot(monthly_sales['Date'], monthly_sales['Sales_Growth'], marker='o', color='red', label='Sales_Growth')

# Menambahkan label untuk data di bar plot
#for bar, value in zip(bars, monthly_sales['Sales_Growth']):
#    plt.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}%', ha='center', va='bottom', fontsize=10, color='black')

plt.title('Pertumbuhan Penjualan dan Total Sales Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Pertumbuhan Penjualan (%)')
plt.tick_params(axis='x', labelsize=10, rotation=90)
plt.legend()
plt.show()
```

There was huge growth in March 2017, sales increased 1132% compared to the previous month's sales. If you look at the sales data in February it was only 4,520, while in March sales increased to 55,691.

### Product Analysis

#### Which product gives the greater profit and order?

```python
#grouping sales by category
category_profit = df.groupby(df['Category'])['Quantity','Profit'].sum()

category_profit
```

```python
category_profit = category_profit.reset_index()

category_profit
```

```python
#count percentage
grotw = category_profit[category_profit['Category']=='Technology']['Profit'].sum()/category_profit['Profit'].sum()

grotw
```

```python
# Bar plot Sales by Category
plt.figure(figsize=(10, 4))

ax1 = plt.subplot(1, 2, 1)
sns.barplot(
    y="Quantity",
    x="Category",
    data=category_profit.sort_values(by="Quantity", ascending=False),
    palette=sns.color_palette(['#38419D', '#52D3D8', '#52D3D8']),
    label='Quantity',
    ax=ax1
)

ax1.set_title("Quantity by Category", loc="center", fontsize=10)
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax1.set_ylabel("Quantity")
ax1.tick_params(axis='x', labelsize=12)

# Bar plot Profit by Category
ax2 = plt.subplot(1, 2, 2)
sns.barplot(
    y="Profit",
    x="Category",
    data=category_profit.sort_values(by="Profit", ascending=False),
    palette=sns.color_palette(['#38419D', '#52D3D8', '#52D3D8']),
    label='Profit',
    ax=ax2
)

ax2.set_title("Profit by Category", loc="center", fontsize=10)
ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
ax2.set_ylabel("Total Profit")
ax2.tick_params(axis='x', labelsize=12)

plt.tight_layout()
plt.show()
```

We can see that the product category that is ordered the most is the Office Supplies category with a total of 22,906 orders or contributing 60% of all products sold. Then looking at profits, it turns out that Technology is the product category with the largest profits with a total profit of 145,455 or 51% of the total profits obtained.
